# Data Analysis

```python
# Imports
import matplotlib as plt
import pandas as pd
import numpy as np
from scipy import stats as st


jurd_df = pd.read_csv("data/Warehouse_and_Retail_Sales.csv")
jurd_df['TOTAL SALES'] = jurd_df['RETAIL SALES'] + jurd_df['WAREHOUSE SALES']
to_drop = ['Item Code', 'Retail Transfers']


jurd_df.head(10)
```

#Basic Analysis: Summary Stats

```python
# Mean
# Mean Retail Sales
avg_retail_sales = np.mean(jurd_df['RETAIL SALES'].to_numpy())
avg_retail_sales = np.round(avg_retail_sales, decimals=2)
print(f'Average Retail Cases Sold: {avg_retail_sales} cases')
# Mean Warehouse Sales
avg_warehouse_sales = np.mean(jurd_df['WAREHOUSE SALES'].to_numpy())
avg_warehouse_sales = np.round(avg_warehouse_sales,decimals=2)
print(f'Average Warehouse Cases Sold: {avg_warehouse_sales} cases')
# Mean Total Sales
avg_total_sales = np.mean(jurd_df['TOTAL SALES'].to_numpy())
avg_total_sales = np.round(avg_total_sales,decimals=2)
print(f'Average Total Cases Sold: {avg_total_sales} cases')

# Median
# Median Retail Sales
median_retail_sales = np.median(jurd_df['RETAIL SALES'].to_numpy())
np_arr = jurd_df['RETAIL SALES'].to_numpy()
sorted_idxs = np.argsort(jurd_df['RETAIL SALES'].to_numpy())
med_idx = sorted_idxs[round(len(sorted_idxs)/2)]
median_retail_sales = np.round(jurd_df['RETAIL SALES'].to_numpy(), decimals=2)
print(f'Median Retail Cases Sold: {np_arr[med_idx]} cases')
# Median Warehouse Sales
median_warehouse_sales = np.median(jurd_df['WAREHOUSE SALES'].to_numpy())
np_arr = jurd_df['WAREHOUSE SALES'].to_numpy()
sorted_idxs = np.argsort(jurd_df['WAREHOUSE SALES'].to_numpy())
med_idx = sorted_idxs[round(len(sorted_idxs)/2)]
median_warehouse_sales = np.round(jurd_df['WAREHOUSE SALES'].to_numpy(), decimals=2)
print(f'Median Warehouse Cases Sold: {np_arr[med_idx]} cases')
# Median Total Sales
median_total_sales = np.median(jurd_df['TOTAL SALES'].to_numpy())
np_arr = jurd_df['TOTAL SALES'].to_numpy()
sorted_idxs = np.argsort(jurd_df['TOTAL SALES'].to_numpy())
med_idx = sorted_idxs[round(len(sorted_idxs)/2)]
median_total_sales = np.round(jurd_df['TOTAL SALES'].to_numpy(), decimals=2)
print(f'Median Total Cases Sold: {np_arr[med_idx]} cases')

# Mode
# Mode Retail Sales
mode_retail_sales = st.mode(jurd_df['RETAIL SALES'].to_numpy())
print(f'Mode Retail Cases Sold: {mode_retail_sales.mode}')
# Mode Warehouse Sales
mode_warehouse_sales = st.mode(jurd_df['WAREHOUSE SALES'].to_numpy())
print(f'Mode Warehouse Cases Sold: {mode_warehouse_sales.mode}')
# Mode Total Sales
mode_total_sales = st.mode(jurd_df['TOTAL SALES'].to_numpy())
print(f'Mode Total Cases Sold: {mode_total_sales.mode}')

# Range
# Retail Range Calculations
retail_sales_max = np.max(jurd_df['RETAIL SALES'].to_numpy()) # Find Max value
retail_sales_min = np.min(jurd_df['RETAIL SALES'].to_numpy()) # Find Min value
range_retail_sales = retail_sales_max - retail_sales_min
print(f'Range of Retail Cases Sold: {range_retail_sales}')
# Warehouse Range Calculations
warehouse_sales_max = np.max(jurd_df['WAREHOUSE SALES'].to_numpy()) # Find Max value
warehouse_sales_min = np.min(jurd_df['WAREHOUSE SALES'].to_numpy()) # Find Min value
range_warehouse_sales = warehouse_sales_max - warehouse_sales_min
print(f'Range of Warehouse Cases Sold: {range_warehouse_sales}')
# Total Range Calculations
total_sales_max = np.max(jurd_df['TOTAL SALES'].to_numpy()) # Find Max value
total_sales_min = np.min(jurd_df['TOTAL SALES'].to_numpy()) # Find Min value
range_total_sales = total_sales_max - total_sales_min
print(f'Range of Total Cases Sold: {range_total_sales}')

# Standard Deviation
# Std of Retail
std_retail_sales = np.std(jurd_df['RETAIL SALES'].to_numpy())
rnd_std_retail_sales = np.round(std_retail_sales, decimals=2)
print(f'Standard Deviation of Retail Cases Sold: {rnd_std_retail_sales} cases')
# Std of Warehouse
std_warehouse_sales = np.std(jurd_df['WAREHOUSE SALES'].to_numpy())
rnd_std_warehouse_sales = np.round(std_warehouse_sales, decimals=2)
print(f'Standard Deviation of Warehouse Cases Sold: {rnd_std_warehouse_sales} cases')
# Std of Total
std_total_sales = np.std(jurd_df['TOTAL SALES'].to_numpy())
rnd_std_total_sales = np.round(std_total_sales, decimals=2)
print(f'Standard Deviation of Total Cases Sold: {rnd_std_total_sales} cases')
```
